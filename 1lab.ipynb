```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

LOWER_LIMIT = 10  # Нижний 
UPPER_LIMIT = np.pi / 3  # Верхний 


def cosine_function(x: float) -> float:

    return np.cos(x)


def main():
    
    # Аналитическое вычисление интеграла
    analytic_result = np.sin(UPPER_LIMIT) - np.sin(LOWER_LIMIT)
    
    # интегрирование
    numeric_result, error_estimate = integrate.quad(
        cosine_function, 
        LOWER_LIMIT, 
        UPPER_LIMIT
    )
    
    # массива точек
    x_values = np.linspace(LOWER_LIMIT - 0.5, UPPER_LIMIT + 0.5, 1000)
    y_values = cosine_function(x_values)
    
    # фигура с графиком
    plt.figure(figsize=(12, 6))
    
    # Построение
    plt.plot(x_values, y_values, 'b-', linewidth=2, label=r'$f(x) = \cos(x)$')
    plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    
    # Закраска
    x_fill = np.linspace(LOWER_LIMIT, UPPER_LIMIT, 100)
    y_fill = cosine_function(x_fill)
    plt.fill_between(
        x_fill, 
        y_fill, 
        alpha=0.3, 
        color='green',
        label=f'Площадь: {numeric_result:.4f}'
    )
    
    # Настройка графика
    plt.grid(True, alpha=0.3)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('cos(x)', fontsize=12)
    plt.title(
        f'Интеграл cos(x) от {LOWER_LIMIT} до {UPPER_LIMIT:.4f}',
        fontsize=14
    )
    plt.legend(fontsize=12)
    plt.xlim(LOWER_LIMIT - 0.5, UPPER_LIMIT + 0.5)
    plt.ylim(-0.2, 1.2)
    
    # Вывод
 
    print("РЕЗУЛЬТАТ ИНТЕГРИРОВАНИЯ")

    print(f"Функция: f(x) = cos(x)")
    print(f"Интервал: [{LOWER_LIMIT}, {UPPER_LIMIT:.4f}]")

    print(f"\nАналитическая формула: ∫cos(x)dx = sin(x)")
    print(f"Результат: sin({UPPER_LIMIT:.4f}) - sin({LOWER_LIMIT}) = {analytic_result:.8f}")
   
    print(f"{'='*151}")
    plt.show()
    print(f"{'='*151}")


if __name__ == "__main__":
    main()
```
