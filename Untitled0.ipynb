```python
from google.colab import drive
drive.mount('/content/drive')

file_path = "/content/drive/MyDrive/my_text_file.txt"
with open(file_path, "w") as f:
    f.write("Это тестовый файл, сохраненный в Google Drive.\n")
    f.write("Вторая строка текста.")

print("Файл успешно сохранен в Google Drive.")

with open(file_path, "r") as f:
  content = f.read()
  print("Содержимое файла:\n", content)

students = [
    ["Прохоров В.И.", 20, "ИВТ-24-2"],
    ["Ганиев Д.И.", 22, "ИВТ-24-2"],
    ["Червиченко И.Д.", 21, "ИВТ-24-2"]
]

csv_path = "/content/drive/MyDrive/stidents.csv"
with open(csv_path, "w") as f:
  for student in students:
    # Corrected to write a comma-separated string for proper CSV format
    f.write(",".join(map(str, student)) + "\n")

print("Файл students.csv успешно сохранен в Google Drive.")

# Now, let's demonstrate loading the CSV file with pandas
import pandas as pd

df_students = pd.read_csv(csv_path, header=None, names=['ФИО', 'Возраст', 'Группа'])
print("\nСодержимое загруженного CSV файла:")
display(df_students)
```

```python
with open(csv_path, 'r', encoding='utf-8') as f:
    content = f.read()

print("\nСодержимое файла:")
print(content)

with open(csv_path, 'a', encoding='utf-8') as f:
    f.write("\n\n# Обработано в Google Colab")

files.download(csv_path)
```
